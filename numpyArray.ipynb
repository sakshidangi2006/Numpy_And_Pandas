```python
"""Deep and shallow copy"""
```

```python
a = np.arange(1,11)
a
```

```python
slice = a[:5]
slice = slice*10          #Shallow copy
slice ,a
```

```python
a = np.arange(1,11)
b = a
b[0] = 77               #Deep copy
a, b
```

```python
a = np.arange(5,16)
b = a.copy()
b[0] = 77
b , a
```
